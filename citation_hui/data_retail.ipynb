```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('amazon_reviews.csv')

# Drop the 'image' column
df = df.drop(columns=['image'])
```

```python
df.info()
```

```python
# Dropping all columns after 'vote'
df_cleaned = df.loc[:, :'vote']
```

```python
df_cleaned
```

```python
df_cleaned.to_csv('clean_retail.csv', index=False)
```

```python
# First, convert the 'vote' column to numeric, coercing errors to NaN
df_cleaned['vote'] = pd.to_numeric(df_cleaned['vote'], errors='coerce')

# Now perform the groupby operation
grouped_df = df_cleaned.groupby(['userName', 'reviewTime']).agg({
    'itemName': 'first',
    'description': 'first',
    'brand': 'first',
    'feature': 'first',
    'category': 'first',
    'price': 'first',
    'rating': 'first',
    'summary': 'first',
    'reviewText': 'first',
    'vote': 'sum'  # This will now work since 'vote' is numeric
}).reset_index()

```
