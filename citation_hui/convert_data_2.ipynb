```python
# Count occurrences of each reference
ref_counts = df['ref'].value_counts()

# Create a new column with the counts
df['n_citation_ref_count'] = df['ref'].map(ref_counts).fillna(0).astype(int)


# Find which rows have changed based on the new column
changed_ids = df[df['n_citation_ref_count'] != df['n_citation_ref']]
```

# Tính toán point theo giá trị đưa vào



```python
import pandas as pd
import numpy as np

# Read the CSV file
file_path = "point_delete_ref0.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)
#df = pd.read_csv(file_path,nrows=1000)
```

```python
# Function to calculate the point
def calculate_point(row, related_refs):
    base_point = 0
    if row['year_ref'] != 0:
        base_point = (row['year_ref'] - 1800) * 10 + row['n_citation_ref'] * 10 + row['n_citation'] * 2
    else:
        base_point = row['n_citation_ref'] * 10 + row['n_citation'] * 2

    # Add 200 points if the ref is in the list of related references
    if row['ref'] in related_refs:
        base_point += 500000

    return base_point

# Example input references that you want to prioritize
related_refs = [2005687710, 2018037215]  # Modify this list based on your input

# Applying the function to calculate points
df['Point'] = df.apply(calculate_point, related_refs=related_refs, axis=1)
```

```python
df.to_csv('citation_input_id.csv',index=False)
```

```python
# Group by 'id' and aggregate the references and points
result = (
    df.groupby('id')
    .agg({
        'ref': lambda x: ' '.join(map(str, x)),  # Use space instead of comma
        'Point': lambda x: f"{x.sum()}:{' '.join(map(str, x))}"  # Format with space
    })
    .reset_index()
)

# Prepare the output format
output_lines = [
    f"{row['ref']}:{row['Point']}" for _, row in result.iterrows()
]

# Save to a .txt file
with open('citation_input_id.txt', 'w') as file:
    for line in output_lines:
        file.write(line + '\n')

print("Data has been saved to 'citation_input_id.txt'.")
```

```python
# Group by 'id' and aggregate the references and points
result = (
    df.groupby('id')
    .agg({
        'ref': lambda x: ','.join(map(str, x)),
        'Point': lambda x: f"{x.sum()}:{','.join(map(str, x))}"
    })
    .reset_index()
)

# Combine the output in the desired format with ID
result['output_with_id'] = result.apply(lambda row: f"{row['id']}:{row['ref']}:{row['Point']}", axis=1)

# Save the output with ID to a .txt file
with open('output_with_id.txt', 'w') as f:
    f.write('\n'.join(result['output_with_id']))

print("Data with IDs has been saved in 'output_with_id.txt'.")
```
