```python
import pandas as pd
import numpy as np

# Read the CSV file
file_path = "id_ref_year_nCitation_docType.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)
#df = pd.read_csv(file_path,nrows=1000)
```

```python
# Count occurrences of each reference
ref_counts = df['ref'].value_counts()

# Create a new column with the counts
df['n_citation_ref_count'] = df['ref'].map(ref_counts).fillna(0).astype(int)


# Find which rows have changed based on the new column
changed_ids = df[df['n_citation_ref_count'] != df['n_citation_ref']]
```

```python
# Update n_citation_ref based on the condition
changed_ids.loc[(df['n_citation_ref'] == 0) & (changed_ids['n_citation_ref_count'] != 0), 'n_citation_ref'] = changed_ids['n_citation_ref_count']
```

```python
changed_ids.loc[changed_ids['ref'] == 0, 'n_citation_ref_count'] = 0
changed_ids.loc[changed_ids['ref'] == 0, 'n_citation_ref'] = 0
```

```python
import pandas as pd
df_delete_ref0 = pd.read_csv('point_delete_ref0.csv')
```

```python
# Current year
current_year = 2024

# Function to calculate citations per year
def calculate_citations_per_year(row):
    if row['year_ref'] > 0:
        return row['n_citation_ref'] / (current_year - row['year_ref'])
    else:
        return 0  # Handle case for year_ref <= 0

# Add the new column to the DataFrame
df_delete_ref0['citations_per_year'] = df_delete_ref0.apply(calculate_citations_per_year, axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
columns_to_scale = ['year', 'year_ref', 'n_citation_ref', 'n_citation', 'citations_per_year']
scaler = MinMaxScaler()
df_delete_ref0[columns_to_scale] = scaler.fit_transform(df_delete_ref0[columns_to_scale])
```

```python
df_delete_ref0
```

```python
# Group by 'id' and prepare to split
grouped = df_delete_ref0.groupby('id')

train_data = []
test_data = []

# Iterate over each group
for name, group in grouped:
    total_refs = len(group)
    
    if total_refs == 1:
        # If there's only one ref, add it to train
        train_data.append(group)
    elif total_refs == 2:
        # If there are two refs, add one to train and one to test
        train_data.append(group.iloc[[0]])  # First ref for training
        test_data.append(group.iloc[[1]])    # Second ref for testing
    else:
        # If there are more than two refs
        train_size = total_refs // 3  # 1/3 for training
        test_size = total_refs - train_size  # 2/3 for testing

        # Assign to train and test
        train_data.append(group.iloc[:train_size])  # First 1/3 for training
        test_data.append(group.iloc[train_size:])   # Remaining 2/3 for testing

# Combine train and test data
train_df = pd.concat(train_data).reset_index(drop=True)
test_df = pd.concat(test_data).reset_index(drop=True)
```
